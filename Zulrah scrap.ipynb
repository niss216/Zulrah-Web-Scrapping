```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import random
import bisect
import collections

#store the url
url='https://oldschool.runescape.wiki/w/Zulrah'

# make the request to the site and access the content
response=requests.get(url)

#extract html from response
page_content=response.content


#parse the pagecontent to get an html page by lxml
soup=BeautifulSoup(page_content,'lxml')

 #find all tables by exploring the inspect in site
my_tables =soup.find_all('table',attrs={"class":"wikitable sortable filterable item-drops autosort=4,a"})

#parse table data


#initialize empty dictionary for iterating the tables and then concat them togather
loot_tables={}
table_index=0

for table in my_tables :
    #parse  out table rows
    table_rows = table.find_all('tr')
   
    
    #parse out table header data
   
    table_header = table.find_all('th')
 

    ## parse columns from table header
    cols=[]
    #for every word in th(header) tag
    for item in table_header :
        if len(item.text)>0:
            cols.append(item.text)
            

    # find the data store in td tag for every table(parse data from each row)
    data=[]
    for row in table_rows:
        table_data = row.find_all('td')
       
        #parse text from the table_data and store it in row_data
        row_data = [item.text for item in  table_data if item.text!= ''] #last condition to avoid blanks
        if len(row_data)>0:
            data.append(row_data)
  
    #Finally store the output as a dataframe by append (cols,data)
    output_frame = pd.DataFrame(data,columns=cols)
    loot_tables.setdefault(table_index , output_frame )
    table_index +=1

       
  
   




```

```python
one_loot_table = pd.concat(loot_tables.values())
one_loot_table.set_index('Item',inplace=True)
one_loot_table 

```
